```python
# here I need numbers, so convert to int
data_list = [int(x) for x in data_list]
```

```python
df = pd.DataFrame({'values': data_list})
```

the question was: of all these measurements, how many are bigger than the previous measurement?  
therefore first calculate the differences, then take the sume of all of those that are larger than zero

```python
differences = np.diff(df['values'].values)
```

```python
# quick sanity check: differences should be one element shorter than the df/original data
len(df) == len(differences) +1
```

```python
sum(differences > 0)
```

then going on to **part 2**:  
it was not informative enough ;-) instead calculate sliding windows of size 3 taking the sum, and calculate from these how many are increasing

```python
# first test with the example data
test_data = [199, 200, 208, 210, 200, 207, 240, 269, 260, 263]
df_test = pd.DataFrame({'test_data': test_data})
```

```python
df_test.test_data.rolling(window=3).sum()
```

```python
# with actual data
df = df.assign(rolling_sum = df['values'].rolling(window=3).sum())
df.head()
```

```python
# assign differences of rolling sum to new column
df = df.assign(diff_rolling = np.nan)
df.loc[3:, 'diff_rolling'] = np.diff(df['rolling_sum'].dropna())
df.head(8)
```

```python
# how many of those are larger than zero?
sum(df['diff_rolling'].dropna() > 0)
```
