```python
def get_array_data(data_list):
    list_of_strings = data_list[0].split(',')
    list_of_ints = [int(x) for x in list_of_strings]
    return list_of_ints
```

```python
int_array = np.array(list_of_ints)
```

```python
def lanternfish_after_1day(fish_age_array):
    fish_age_array = fish_age_array -1
    spawning_indices = np.where(fish_age_array < 0)[0]
    fish_age_array[spawning_indices] = 6
    fish_age_array = np.append(fish_age_array, len(spawning_indices) * [8])
    return fish_age_array
```

```python
n_days = 80
while n_days > 0:
    int_array = lanternfish_after_1day(int_array)
    n_days -= 1
```

```python
int_array = np.array(list_of_ints)
```

```python
# build my initial count of fish as numpy array
count = np.empty(9)
for ind in np.arange(9):
    count[ind] = np.count_nonzero(int_array == ind)
count
```

```python
def grouped_fish_after_1day(count):
    # how many will be spawning today
    spawning = count[0]
    # all get one day closer to spawning
    count[0:8] = count[1:9]
    # add new ones
    count[8] = spawning
    # keep the old ones in as well
    count[6] += spawning
    return count
```

```python
n_days = 256
while n_days > 0:
    count = grouped_fish_after_1day(count)
    n_days -= 1
```
